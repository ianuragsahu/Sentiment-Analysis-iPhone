### Data Preprocessing

This code is an automated data preprocessing pipeline for a collection of text comments

* Loading the Data: The code starts by loading a dataset. Each file represents a source of comments, like discussion threads or YouTube video comments. 

* Removing Unnecessary Information: In many datasets, there's often metadata or header information at the beginning that doesn't contain actual comments. This code removes such metadata, streamlining the dataset to include only the comments themselves. It's akin to discarding unnecessary packaging from a product.

* Text Cleaning: This is where the code tackles the comments themselves:

* Lowercasing: All text is converted to lowercase to ensure uniformity. This is vital because "Apple" and "apple" should be treated as the same word.
* Punctuation Removal: Punctuation marks (e.g., '.', ',', '!') are stripped away. These symbols don't contribute to the text's meaning in the context of advanced analysis.
* Emoji Removal: Emojis are extracted from the text. While emojis add emotion and context to text, they can also complicate text analysis, so they are isolated for separate consideration.
* Tokenization: The comments are divided into individual words or tokens. Tokenization breaks down complex text into digestible chunks for analysis.
* Removing Stop Words: Common words like "and," "the," and "in" – known as stop words – are removed. They're considered noise in text analysis because they're frequent but often lack meaningful content.

* Data Transformation: The cleaned text data is structured into a format suitable for analysis.

This code ensures that the text data is in a "clean" and "tidy" state, making it ready for the application of natural language processing (NLP) and machine learning techniques.


```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Load the CSV file
file_name = "cleaned_and_merged_reviews_iphone14.csv"
df = pd.read_csv(file_name)

# Remove the header (first row) from the DataFrame
df = df.iloc[1:]

# Function to preprocess text
def preprocess_text(text):
    # Check if the text is a NaN value
    if pd.isna(text):
        return ""  # Return an empty string for NaN values
    
    # Remove unnecessary emojis
    text = re.sub(r'[\U00010000-\U0010ffff]', '', text)
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    
    # Tokenize the text into words
    words = nltk.word_tokenize(text)
    
    # Remove stop words
    words = [word for word in words if word not in stopwords.words('english')]
    
    # Join the cleaned words back into a sentence
    cleaned_text = ' '.join(words)
    
    return cleaned_text

# Apply text preprocessing to the 'reviews' column
df['reviews'] = df['reviews'].apply(preprocess_text)

df.to_csv(file_name, index=False)

print(f"Preprocessed data saved back to '{file_name}' without the header and column name.")
```

```python
# Function to extract keyword frequencies using the "reviews" column
def extract_keyword_frequencies_updated(data, keywords):
    keyword_frequencies = {}
    for keyword in keywords:
        keyword_frequencies[keyword] = data['reviews'].str.contains(keyword, case=False, na=False).sum()
    return keyword_frequencies

# Extract keyword frequencies for both datasets
iphone_14_keywords_freq_updated = extract_keyword_frequencies_updated(iphone_14_data, keywords)
iphone_15_keywords_freq_updated = extract_keyword_frequencies_updated(iphone_15_data, keywords)

iphone_14_keywords_freq_updated, iphone_15_keywords_freq_updated
```

* From the data presented, we can observe a notable difference in the frequency of certain keywords associated with the iPhone 14 and iPhone 15. Specifically, the term "battery" is mentioned 592 times when discussing the iPhone 14, while it has a slightly higher occurrence of 731 mentions for the iPhone 15. This could indicate a greater emphasis or discussion around the battery for the latter model. 

* On the other hand, the keyword "titanium" shows a stark contrast between the two models. It's mentioned a mere once for the iPhone 14, suggesting it might not have been a significant feature or topic of discussion for that model. However, for the iPhone 15, "titanium" has been referenced a substantial 542 times, indicating that it might be a prominent or newsworthy feature for this particular model.

These frequency counts help in understanding which keywords are mentioned most frequently in the reviews for each iPhone model. This information can be valuable for sentiment analysis and topic modeling to gain insights into customer perceptions and preferences.

### Sentiment analysis - VADER (Valence Aware Dictionary and sEntiment Reasoner)

The code employs the VADER (Valence Aware Dictionary and sEntiment Reasoner) sentiment intensity analyzer, a tool specifically designed to assess sentiment in text data. The primary objective is to analyze the sentiments expressed in reviews for both iPhone 14 and iPhone 15.

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize VADER sentiment intensity analyzer
sia = SentimentIntensityAnalyzer()

# Function to categorize sentiment based on compound score
def categorize_sentiment(score):
    if score > 0.05:
        return "positive"
    elif score < -0.05:
        return "negative"
    else:
        return "neutral"

# Function to perform sentiment analysis on a dataset
def perform_sentiment_analysis(data, column_name):
    # Get sentiment scores
    data['compound'] = data[column_name].apply(lambda x: sia.polarity_scores(str(x))['compound'])
    # Categorize sentiment based on compound score
    data['sentiment'] = data['compound'].apply(categorize_sentiment)
    # Calculate the proportion of each sentiment category
    sentiment_distribution = data['sentiment'].value_counts(normalize=True) * 100
    return sentiment_distribution

# Perform sentiment analysis on both datasets
iphone_14_sentiments = perform_sentiment_analysis(iphone_14_data, 'reviews')
iphone_15_sentiments = perform_sentiment_analysis(iphone_15_data, 'reviews')

# Print the sentiment distribution for iPhone 14 and iPhone 15
print("Sentiment Distribution for iPhone 14:\n", iphone_14_sentiments)
print("\nSentiment Distribution for iPhone 15:\n", iphone_15_sentiments)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd

sia = SentimentIntensityAnalyzer()

# List of keywords related to functionalities and features
features_keywords = [
    "battery", "camera", "performance", "screen", "ios", "innovation", 
    "charging", "storage", "design", "price", "software",
    "security", "connectivity", "audio quality", "build quality", 
    "user interface", "compatibility", "updates", "accessibility", 
    "customization", "camera quality", "battery life", "speed", 
    "memory", "screen resolution", "durability"
]

# Function to perform sentiment analysis for features in both iPhone 14 and iPhone 15 reviews
def sentiment_analysis_for_features(data, column_name, features_keywords):
    feature_sentiments = {feature: "neutral" for feature in features_keywords}
    for _, row in data.iterrows():
        review = str(row[column_name])
        for feature in features_keywords:
            # Check if the review contains the feature keyword
            if feature in review:
                sentiment_score = sia.polarity_scores(review)['compound']
                # Classify sentiment as positive, negative, or neutral based on compound score
                if sentiment_score > 0.05:
                    feature_sentiments[feature] = "positive"
                elif sentiment_score < -0.05:
                    feature_sentiments[feature] = "negative"
    return feature_sentiments

# Perform sentiment analysis for features in both iPhone 14 and iPhone 15 reviews
iphone_14_feature_sentiments = sentiment_analysis_for_features(iphone_14_data, 'reviews', features_keywords)
iphone_15_feature_sentiments = sentiment_analysis_for_features(iphone_15_data, 'reviews', features_keywords)

# Create dataframes from the sentiment analysis results
df_14 = pd.DataFrame(iphone_14_feature_sentiments.items(), columns=['Feature', 'Sentiment'])
df_15 = pd.DataFrame(iphone_15_feature_sentiments.items(), columns=['Feature', 'Sentiment'])

# Merge df_14 and df_15 based on the 'Feature' column
merged_df = df_14.merge(df_15, on='Feature', suffixes=('_iPhone_14', '_iPhone_15'))

# Rename the columns to indicate the source (iPhone 14 or iPhone 15)
merged_df.rename(columns={'Sentiment_iPhone_14': 'Sentiment iPhone 14', 'Sentiment_iPhone_15': 'Sentiment iPhone 15'}, inplace=True)

# Display the merged table
print("Merged Sentiment Analysis Results for iPhone 14 and iPhone 15 Features:")
print(merged_df)
```

## LDA -  Latent Dirichlet Allocation

The objective of using Latent Dirichlet Allocation (LDA) is to automatically identify and extract meaningful topics or themes that are being discussed by reviewers without having to manually read and categorize each review. By applying LDA, we can uncover common subjects, such as camera quality, battery life, design, and other aspects related to iPhones, from a large set of reviews. This helps in gaining insights into what customers are discussing and what aspects of the iPhone are most prominent in their reviews.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Function to extract topics using Latent Dirichlet Allocation (LDA)
def extract_topics(data, column_name, n_topics=5, n_words=5):

    vectorizer = CountVectorizer(stop_words='english', max_df=0.95, min_df=2)
    dtm = vectorizer.fit_transform(data[column_name].astype(str))

    lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
    lda.fit(dtm)

    topics = {}
    for index, topic in enumerate(lda.components_):
        # Retrieve the top 'n_words' words for each topic
        topics[index] = [vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-n_words:]]
    return topics

# Extract topics for both datasets (presumably iPhone 14 and iPhone 15 reviews)
iphone_14_topics = extract_topics(iphone_14_data, 'reviews')
iphone_15_topics = extract_topics(iphone_15_data, 'reviews')

# Print the extracted topics for analysis
print("Topics for iPhone 14 Reviews:")
print(iphone_14_topics)
print("\nTopics for iPhone 15 Reviews:")
print(iphone_15_topics)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import pandas as pd

# 1. Generate BERT embeddings for reviews
model = SentenceTransformer('distilbert-base-nli-mean-tokens')
iphone_14_embeddings = model.encode(iphone_14_data['reviews'].astype(str).tolist(), show_progress_bar=True)
iphone_15_embeddings = model.encode(iphone_15_data['reviews'].astype(str).tolist(), show_progress_bar=True)

# 2. Reduce dimensionality
pca = PCA(n_components=10)  # Reducing to 10 components as an example
iphone_14_embeddings_pca = pca.fit_transform(iphone_14_embeddings)
iphone_15_embeddings_pca = pca.fit_transform(iphone_15_embeddings)

# 3. Cluster the reduced embeddings (using KMeans as an example)
n_clusters = 5  # You chose 5 topics for LDA, so we'll stick with that for comparison
kmeans_14 = KMeans(n_clusters=n_clusters).fit(iphone_14_embeddings_pca)
kmeans_15 = KMeans(n_clusters=n_clusters).fit(iphone_15_embeddings_pca)

# Add cluster labels to your dataframe
iphone_14_data['topic'] = kmeans_14.labels_
iphone_15_data['topic'] = kmeans_15.labels_

# 4. Interpret the clusters
def get_representative_reviews(data, n=5):
    topics = {}
    for cluster_num in range(n_clusters):
        representative_reviews = data[data['topic'] == cluster_num]['reviews'].values[:n]
        topics[cluster_num] = representative_reviews
    return topics

iphone_14_topics_bert = get_representative_reviews(iphone_14_data)
iphone_15_topics_bert = get_representative_reviews(iphone_15_data)

print("Topics for iPhone 14 Reviews using BERT:")
print(iphone_14_topics_bert)
print("\nTopics for iPhone 15 Reviews using BERT:")
print(iphone_15_topics_bert)
```
